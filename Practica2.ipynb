```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
Fabrica1=[]
for _ in range(100):
  numero_aleatorio=round(random.uniform(50,1000),2)
  Fabrica1.append(numero_aleatorio)

print(Fabrica1)
```

```python
Fabrica2=[]
for _ in range(100):
  numero_aleatorio=round(random.uniform(50,1000),2)
  Fabrica2.append(numero_aleatorio)

print(Fabrica2)
```

#MAXIMO, MINIMO, RANGO

```python
maximo1=max(Fabrica1)
minimo1=min(Fabrica1)
rango1=maximo1-minimo1
media1=mean(Fabrica1)
mediana1=median(Fabrica1)
moda1= mode(Fabrica1)
desviacion_estandar_f1=stdev(Fabrica1)
print("Estadística descriptiva de la Fabrica 1 \n","Media=",media1,
      "\n Mediana=",mediana1,"\n Moda=",moda1,"\n Máximo=",maximo1,
      "\n Minimo=", minimo1,"\n Rango=", rango1)
print("Desviación Estandar=", desviacion_estandar_f1)
```

```python
maximo2=max(Fabrica2)
minimo2=min(Fabrica2)
rango2=maximo2-minimo2
media2=mean(Fabrica2)
mediana2=median(Fabrica2)
moda2= mode(Fabrica2)
desviacion_estandar_f2=stdev(Fabrica2)
print("Estadística descriptiva de la Fabrica 2 \n","Media=",media2,
"\n Mediana=",mediana2,"\n Moda=",moda2,"\n Máximo=",maximo2,
"\n Minimo=", minimo2,"\n Rango=", rango2)
print("Desviación Estandar=", desviacion_estandar_f2)
print("Error: La lista Fabrica2 está vacía. Por favor, asegúrese de que la lista se haya poblado correctamente en la celda anterior.")
```

```python
plt.figure(figsize=(10,6))
plt.hist(Fabrica2, bins=10, alpha=0.5, color='orange', edgecolor="black", label="Fabrica 2")
plt.axvline(media2, color='orange', linestyle="dashed", linewidth=2, label=f"Media: {media2:.2f}")
plt.axvline(mediana2, color='purple', linestyle="dashed", linewidth=2, label=f"Mediana: {mediana2:.2f}")
plt.axvline(moda2, color='brown', linestyle="dashed", linewidth=2, label=f"Moda: {moda2:.2f}")
plt.axvline(desviacion_estandar_f2, color='lime', linestyle="dashed", linewidth=2, label=f"Desv. Estándar: {desviacion_estandar_f2:.2f}")
plt.title("Histograma de Fabrica 2 con Media, Mediana, Moda y Desviación Estándar")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```
