```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,6,7,8,9,10])
p=np.array([0.05,0.10,0.20,0.30,0.25,0.10])
print("SUma de probabilidades:", np.sum(p))
E=np.sum(x*p)
print("valor esperado:", E)
E2=np.sum((x**2)*p)
var=E2-E**2
print("Varianza:", var)
std=np.sqrt(var)
print("Desviacion estandar:", std)
F=np.cumsum(p)
print("F(x):",F)

plt.step(x,F,where='post')
plt.scatter(x,F)
plt.xlabel("Premio ($)")
plt.ylabel("F(X)")
plt.title("Funcion de distribucion acumulada")
plt.grid()
plt.show()
```
