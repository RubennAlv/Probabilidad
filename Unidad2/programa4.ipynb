```python
import math

# Definición de variables según el ejercicio
n = 10  # Total de asistentes disponibles
k = 4   # Preguntas a calificar (subconjunto)

def calcular_permutaciones(n, k):
    """
    Calcula el número de permutaciones P(n, k) = n! / (n - k)!
    """
    # Método 1: Usando la fórmula matemática directamente
    permutaciones = math.factorial(n) // math.factorial(n - k)
    return permutaciones

# Ejecución y visualización de resultados
resultado = calcular_permutaciones(n, k)

print(f"Total de asistentes (n): {n}")
print(f"Preguntas a calificar (k): {k}")
print(f"El número de maneras (permutaciones) es: {resultado}")

# Método 2: Usando la función especializada de la librería math (disponible en Python 3.8+)
print(f"\nVerificación con math.perm: {math.perm(n, k)}")
```
