```python
import math as mt
from itertools import *
import numpy as np
caracteres = ["A","B","C","D","E","F","0","1","2","3","4","5"]
n=len(caracteres)
r=10 #¿como agrupar?
Cn = mt.factorial(n) / (mt.factorial(r) * mt.factorial(n-r))
print("Exixsten ", Cn, " posibles combinaciones sin repeticion del total de ", n, " caracteres en grupos de ", r, " en ", r)
print("Combinaciones:")
combinaciones_caracteres = list(combinations(caracteres, r))
for combinacion in combinaciones_caracteres:
  print(combinacion)
```
