```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Datos
mu = 70
sigma = 4
n = 500
x1, x2 = 68, 72

# Probabilidad
prob = norm.cdf(x2, mu, sigma) - norm.cdf(x1, mu, sigma)
estudiantes = prob * n

print("Probabilidad:", round(prob, 4))
print("Porcentaje:", round(prob*100, 2), "%")
print("Número de estudiantes:", round(estudiantes))

# Rango de valores
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Gráfica
plt.figure()
plt.plot(x, y)

# Área sombreada
x_fill = np.linspace(x1, x2, 1000)
y_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill)

# Líneas verticales
plt.axvline(x1)
plt.axvline(x2)


# Títulos
plt.title("Distribución Normal (μ=70, σ=4)")
plt.xlabel("Peso (kg)")
plt.ylabel("Densidad")

plt.show()
```
