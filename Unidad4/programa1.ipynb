```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- 1. Parámetros y Generación de Datos (Pág 13) ---
media = 70
desv_std = 4
num_datos = 500

# Generar la muestra aleatoria
datos = np.random.normal(media, desv_std, num_datos)

# --- 2. Crear la gráfica uniendo Histograma y Densidad (Págs 14 a 17) ---

# Dibujar el histograma normalizado (density=True es clave para que empate con la curva)
plt.hist(datos, bins=30, density=True, color='skyblue', edgecolor='black', label='Frecuencia (Histograma)')

# Calcular los datos para la curva de densidad (PDF)
xmin, xmax = plt.xlim()             # Obtenemos los límites del eje X que dejó el histograma
x = np.linspace(xmin, xmax, 100)    # Generamos 100 puntos distribuidos en ese rango
p = norm.pdf(x, media, desv_std)    # Calculamos la probabilidad para cada uno de esos puntos

# Superponer la curva de densidad en color rojo
plt.plot(x, p, 'k', linewidth=2, color='red', label='Distribución (PDF)')

# Formato final de la gráfica
plt.title('Distribución Normal (Media=70, Desv=4)')
plt.legend()

# Mostrar todo junto
plt.show()
```
